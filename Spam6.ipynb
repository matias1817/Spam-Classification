```python
from nltk.tokenize import word_tokenize
```

```python
def tokenizar_texto(texto):
    return nltk.word_tokenize(str(texto))
```

```python
import nltk
nltk.download('punkt')
```

```python
texto_token = emails.apply(tokenizar_texto)
texto_token
```

```python
vectorizer = CountVectorizer()
```

```python
#strig_tokenizada = [' '.join(tokens_doc) for tokens_doc in texto_toke]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
string_filtrada = []
for i in range(texto_token.size):
  string_filtrada.append([w for w in texto_token.iloc[i] if not w.lower() in stop_words])
```

```python
print(string_filtrada)
```

```python
from nltk.stem import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
string_stemizada = []
for i in range(len(string_filtrada)):
  string_stemizada.append([stemmer.stem(w) for w in string_filtrada[i]])
```

```python
 for i in range(len(string_stemizada)):
  string_stemizada[i].append(str(len(string_stemizada[i])))
```

```python
print(string_stemizada)
```

```python
#vetores_emails = vectorizer.fit_transform(string_stemizada)
vetores_emails = vectorizer.fit_transform([' '.join(inner_list).lower() for inner_list in string_stemizada])
```

```python
print(vetores_emails)
```

```python
vetores_emails_train, vetores_emails_test, spam_ham_train, spam_ham_test = train_test_split(vetores_emails, spam_ham, test_size=0.2, random_state=42, stratify=spam_ham)
```
